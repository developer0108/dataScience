```python
# Importing the necessary packages
import numpy as np                                  # "Scientific computing"
import pandas as pd                                 # Data Frame



```

```python
# Load the XLSX file into a DataFrame
df = pd.read_excel('https://github.com/HoGentTIN/dsai-labs/raw/main/data/belgian_population.xlsx')
```

The file belgian_population.xlsx contains data about the Belgian population by marital status, age and gender in 2023. There are 4 marital statuses:  
- unmarried
- married
- widowed
- divorced

```python
df.head()
```

1. We want to add a separate column showing the total number of women and the total Number_of_Men by age.  
Before we can sum, however, we need to replace all NaNs with 0. Otherwise, the total sum by age will also be NaN. We are nothing with that.

```python
df = df.fillna(0)
df.head()
```

2. Add two new columns:
- the total number of women by age
- the total Number_of_Men by age



```python
df['Number_of_Women'] = df['Women_Unmarried'] + df['Women_Widowed'] + df['Women_Married'] + df['Women_Divorced']
df['Number_of_Men'] = df['Men_Unmarried'] + df['Men_Widowed'] + df['Men_Married'] + df['Men_Divorced']
df.head(50)
```

3. What is the total number of women and men in Belgium?

```python
print(f'total number of men in Belgium: {sum(df["Number_of_Men"])}')
print(f'total number of women in Belgium: {sum(df["Number_of_Women"])}')
```

4. What is the average age for women and men in Belgium? Do women still outlive men?

```python
malePopulation = sum(df['Number_of_Men'])
femalePopulation = sum(df['Number_of_Women'])

maleAges = df[['Age', 'Number_of_Men']]
femaleAges = df[['Age', 'Number_of_Women']]

maleAges['Years_lived'] = maleAges['Age'] * maleAges['Number_of_Men']
femaleAges['Years_lived'] = femaleAges['Age'] * femaleAges['Number_of_Women']

maleAverageAge = sum(maleAges['Years_lived']) / malePopulation
femaleAverageAge = sum(femaleAges['Years_lived']) / femalePopulation

print(maleAverageAge, femaleAverageAge)
if maleAverageAge > femaleAverageAge :
    print('males have a higher average age than females')
else:
    print('Females have a higher average age than males')
```

5. A first look at the data shows, that the distribution of girls vs boys is not 50/50. This is a well-known scientific phenomenon.  
Calculate the total number of girls through age 5, and the total number of boys through age 5. Based on this, calculate the exact distribution of girls vs boys.

Percentage of girls = 0.48888139696591654
Percentage of boys = 0.5111186030340835
6

```python
totalWomenFrom5 = sum(df.loc[df['Age'] <= 5]['Number_of_Women'])
totalMenFrom5 = sum(df.loc[df['Age'] <= 5]['Number_of_Men'])
totalPopulation = totalMenFrom5 + totalWomenFrom5

print(f'percentage of girls = {totalWomenFrom5 / totalPopulation}')
print(f'percentage of boys =  {totalMenFrom5 / totalPopulation}')
```

6. At birth, there are more boys than girls. However, boys die younger, partly due to reckless behavior.  
From what age do women outnumber men?

```python
print(min(df.loc[df['Number_of_Women'] > df['Number_of_Men']]['Age']))
```

```python
for x in range(15, 56, 5):
    womenMarriedForPercentage = min(df.loc[(df['Women_Married'] / df['Number_of_Women']) * 100 > x]['Age'])
    menMarriedForPercentage = min(df.loc[(df['Men_Married'] / df['Number_of_Men']) * 100 > x]['Age'])

    print(f'Age at which {x}% of the women are married = {womenMarriedForPercentage}')
    print(f'Age at which {x}% of the men are married = {menMarriedForPercentage}')
    print('\n')

```
