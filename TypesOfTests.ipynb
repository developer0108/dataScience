## Measure of Central Tendency

Welke waarde representateert de volledige groep?

### Arithmetic mean (Gemiddelde)

De som van alle values gedeeld door het aantal values

### Median

Het midden van alle values na het sorteren (bij even getallen, midden van de twee middelste)

### Mode

De value die het meest voorkomt in de dataset



## Measures of Dispersion

### Range = bereik

De absolute waarde van het verschil tussen de hoogste en laagste value

### Quartiles = kwartielen

De drie waardes die een gesorteerde dataset verdelen in 4 gelijke subsets (Q1,Q2,Q3)

### Interquartile Range = interkwartielafstand

Het verschil tussen de derde en eerste quartile | Q3 - Q1 |

### Variance s² = variantie

Variantie zegt iets over de spreiding van een dataset. Hoe meer verspreidt de data is, hoe groter de variantie ten opzichte van het gemiddelde. Het is de mate waarin een aantal waarden van elkaar veschillen

(s²) = Σ [(xi - x̅)²]/n - 1

s² = Variantie  
Σ = Sommatie, de som van alle termen in de vergelijking na het sommatieteken.  
xi = De steekproefwaarden.  
x̅ = Het gemiddelde van de waarden in de reeks.  
n = De steekproefgrootte. Het aantal waarden in de reeks.  


### Standard deviation

De standaard deviatie is de vierkantswortel van de variatie

De standaard deviatie geeft de mate van spreiding in bepaalde data. Het geeft aan hoeveel de geobserveerde waardes afwijken van het gemiddelde.

## Testing Procedure

### Statistical Hypothesis Testing

Hypothese = idee dat nog bewezen moet worden: stelling over de numerieke waarde van de populatie parameter
Hypthese Test = verificatie van de stelling over de waarde van één of meerdere populatie parameters
Null Hypothesis (H0) = basis hypothese waar verondersteld wordt dat ze waar is
Alternatieve hypothese (H1,Ha) = conclusie of de nul hypothese onwaar kan zijn

### Elements of a testing procedure

Test Statistic = de waarde die berekend wordt via de sample
Region of Acceptance = de regio van waardes die de null hypothese ondersteuning
Critical Region / Region of Rejection = de regio van waardes die de null hypothese afwijzen
Significance Level (a)= de kans om een null hypothese te verwerpen

### Probability Value

p-value = de kans dat, als de nulll hypothese waar is, een waarde te krijgen voor de test statistic dat minstens zo extreem is als de observed value

### Critical Region

Collectie van alle waarden van de test statistic voor welke de null hypothesis verworpen kan worden

Je wilt weten of de gemiddelde sprintsnelheid van kinderen uit groep 7 afwijkt van die van kinderen uit groep 8.  
 
## Two-sample t-test

Gebruik de paired samples t-test om twee gemiddelden van gepaarde steekproeven met elkaar te vergelijken. Gepaarde steekproeven zijn afhankelijk van elkaar.  

Je meet de lengte van dezelfde personen in 2015 en 2018. Deze waarden zijn afhankelijk van elkaar (omdat je dezelfde persoon meet), en daarom gebruik je een paired samples t-test.  

# Voorwaarden

Voordat je een t-test uitvoert, is het belangrijk om ervoor te zorgen dat je data aan een aantal voorwaarden voldoen. Pas als aan alle voorwaarden is voldaan, zijn de resultaten van de t-test betrouwbaar.

De afhankelijke variabele wordt gemeten op interval- of rationiveau (scale).  
De personen binnen de twee groepen zijn onafhankelijk van elkaar. Je kunt hiervoor werken met aselecte steekproeven.  
Als je steekproef minder dan 30 observaties telt, moet de afhankelijke variabele normaal verdeeld zijn. Dit kun je controleren in SPSS met de Shapiro-Wilk- of Kolmogorov-Smirnov-toets. Als de variabele niet normaal verdeeld is, kun je beter de Wilcoxon- of de Mann-Whitney-toets gebruiken.

## Linear Regression

With regression we will try to find a consistent and systematic relationship between two qualitative variables.  

1. Monotonic:consistent direction of the relationship between the twovariables: increasing or decreasing  
2. Non-monotonic:value of dependent variable changessystematically with value of independent variable, but the directionis not consistent  

### Covariance

Toont aan of de relatie tussen twee variabelen stijgt of daalt  

Cov > 0: increasing  
Cov ≈ 0: no relationship  
Cov < 0: decreasing  

### Pearson correlation coefficient

Meet de sterkste van de lineaire correlatie tussen x en y  

The correlation coefficient only looks at the relationship between twovariables. Interactions with other variables are not considered.  
The correlation coefficient explicitly does not assume a causalrelationship.  

De Pearson correlatiecoëfficiënt (Pearson correlation coefficient), aangeduid met r, is de meest gebruikelijke manier om een lineaire correlatie te meten.  
Het is een getal tussen de -1 en 1 dat de sterkte en de richting van het verband tussen twee variabelen meet.



### Coefficient of determination

The coefficient of determination 𝑅² explains the percentage of the variance of the observed values relative to the regression line.
Pearson’s correlation coefficient only expresses linear relationships.
