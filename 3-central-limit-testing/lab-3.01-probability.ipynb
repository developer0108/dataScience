## Exercise 1.2 - Plot probability density

Make a graphical plot of the probability density function and the cumulative distribution function of a normal distribution with $\mu = 2.5$ and $\sigma = 1.5$.

What is the area under the density function between  x=0.5  and  x=4 . (Answer = 0.750)

```python
m = 2.5
s = 1.5

x_axis = np.linspace(m * s - 5 , m * s + 5 , 202)
y_axis = np.linspace(m * s - 5 , m * s + 5, 202)

x = stats.norm.pdf(x_axis, m, s)
y = stats.norm.cdf(y_axis, m, s)
plt.plot(x_axis, x)
plt.plot(y_axis, y)

plt.fill_between(x_axis, 0, x ,where=(x_axis >= 0.5) & (x_axis <=4) , color='b')
plt.fill_between(y_axis, 0, y ,where=(y_axis >= 0.5) & (y_axis <=4) , color='r')
plt.axvline(0.5)
plt.axvline(4)
# dist_x, 0, dist_y, where=dist_x <= x, color='lightblue')
plt.show()
print(f"{stats.norm.cdf(4, m, s) - stats.norm.cdf(0.5, m, s):.2f}")

```

## Exercise 1.3 - Student t vs normal distribution

Make a plot of the probability density function and the cumulative distribution function for the Student's t-distribution with 3 degrees of freedom. 

Also plot the standard normal distribution, so you can compare both.

```python
m = 2.5
s = 1.5

x_axis = np.linspace(m * s - 5 , m * s + 5 ,202)
y_axis = np.linspace(m * s - 5 , m * s + 5, 202)

x = stats.t.pdf(x_axis, df=3)
y = stats.t.cdf(y_axis, df=3)
plt.plot(x_axis, x)
plt.plot(y_axis, y)

plt.fill_between(x_axis, 0, x ,where=(x_axis >= 0.5) & (x_axis <=4) , color='b')
plt.fill_between(y_axis, 0, y ,where=(y_axis >= 0.5) & (y_axis <=4) , color='r')
plt.axvline(0.5)
plt.axvline(4)
# dist_x, 0, dist_y, where=dist_x <= x, color='lightblue')
plt.show()
print(f"{stats.t.cdf(4, df=3) - stats.t.cdf(0.5, df=3):.2f}")
```
